```python
import time
import matplotlib.pyplot as plt

def time_it(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter_ns()
        result = func(*args, **kwargs)
        end = time.perf_counter_ns()
        return (end - start), result
    return wrapper

```

# First algoritm

```python
@time_it
def first(n):
    # initialize all values in c to True
    c = [True]*(n+1)
    c[0] = False
    c[1] = False
    
    i = 2
    while i <= n:
        if c[i]:
            j = 2*i
            while j <= n:
                c[j] = False
                j += i
        i += 1
    
    # return a list of all primes up to n
    primes = []
    for i in range(2, n+1):
        if c[i]:
            primes.append(i)
    
    return primes

print(first(100)[1])
```

# Second algoritm

```python
@time_it
def second(n):
    # initialize all values in c to True
    c = [True]*(n+1)
    c[0] = False
    c[1] = False
    
    i = 2
    while i <= n:
        j = 2*i
        while j <= n:
            c[j] = False
            j += i
        i += 1
    
    # return a list of all composite numbers up to n
    composites = []
    for i in range(2, n+1):
        if c[i]:
            composites.append(i)
    
    return composites
print(second(100)[1])
```

# Third algormitm

```python
@time_it
def third(n):
    # initialize all values in c to True
    c = [True]*(n+1)
    c[0] = False
    c[1] = False
    i = 2
    while i <= n:
        if c[i]: 
            j = i + 1 
            while j <= n:
                if j % i == 0:
                    c[j] = False
                j += 1
        i += 1
    
    # return a list of all composite numbers up to n
    composites = []
    for i in range(2, n+1):
        if c[i]:
            composites.append(i)
    
    return composites
print(third(100)[1])
```

# Fourth algoritm

```python
@time_it
def fourth(n):
    # initialize all values in c to True
    c = [True]*(n+1)
    c[0] = False
    c[1] = False
    
    i = 2
    while i <= n:
        j = 2
        while j < i:
            if i % j == 0:
                c[i] = False
            j += 1
        i += 1
    
    # return a list of all primes up to n
    primes = []
    for i in range(2, n+1):
        if c[i]:
            primes.append(i)
    
    return primes
print(fourth(100)[1])
```

# Fifth algoritm

```python
from math import sqrt

@time_it
def fifth(n):
    # initialize all values in c to True
    c = [True]*(n+1)
    c[0] = False
    c[1] = False
    
    i = 2
    while i <= n:
        j = 2
        while j <= int(sqrt(i)):
            if i % j == 0:
                c[i] = False
            j += 1
        i += 1
    
    # return a list of all primes up to n
    primes = []
    for i in range(2, n+1):
        if c[i]:
            primes.append(i)
    
    return primes

print(fifth(100)[1])
```

```python
def plot_m(funcs):
    for func in funcs:
        values = []
        for i in range(1000, 1200):
            values.append(func[0](i)[0])
        
        plt.plot([i for i in range(1000, 1200)], values, label=func[1])
    plt.legend()
    plt.show()

plot_m([(first, "first"), (second, "second"), (third, 'third'), (fourth, 'fourth'), (fifth, 'fifth')])
plot_m([(first, "first"), (second, "second")])
```
